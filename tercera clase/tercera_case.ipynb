## Serie de tiempo estocastico

Las series de tiempo estocasticas son aquellas que presentan un comportamiento aleatorio o impredecible. A diferencia de las series de tiempo determinísticas, donde los valores futuros pueden ser predichos con certeza a partir de los valores pasados, las series de tiempo estocásticas contienen un componente de ruido o aleatoriedad que dificulta la predicción precisa.

### Estacionariedad

Una serie de tiempo estocástica se considera estacionaria si sus propiedades estadísticas, como la media y la varianza, son constantes a lo largo del tiempo. Esto significa que la serie no muestra tendencias o patrones estacionales y que las fluctuaciones alrededor de la media son consistentes.

```python
# vamos a ver un ejemplo con un grafico 
import numpy as np
import matplotlib.pyplot as plt
# Generar una serie de tiempo estocástica
np.random.seed(0)  # Para reproducibilidad
n = 100
time = np.arange(n)
# Crear una serie de tiempo con una tendencia y ruido aleatorio
trend = 0.5 * time
noise = np.random.normal(0, 5, n)
series = trend + noise
# Graficar la serie de tiempo
plt.figure(figsize=(10, 5))
plt.plot(time, series, label='Serie de Tiempo Estocástica')
plt.title('Ejemplo de Serie de Tiempo Estocástica')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()
plt.grid()
plt.show()
```

Podemos decir que si el promedio depende del tiempo, entonces la serie no es estacionaria porque el promedio no es constante. De igual forma, si la varianza depende del tiempo, entonces la serie no es estacionaria porque la varianza no es constante. Por último, si la covarianza entre dos periodos de tiempo depende del tiempo, entonces la serie no es estacionaria porque la covarianza no es constante.

```python
# ejemplo de una serie de tiempo estacionaria
# Generar una serie de tiempo estacionaria
np.random.seed(0)  # Para reproducibilidad
n = 100
time = np.arange(n)
# Crear una serie de tiempo estacionaria con media constante y ruido aleatorio
mean = 10
noise = np.random.normal(0, 5, n)
stationary_series = mean + noise
# Graficar la serie de tiempo estacionaria
plt.figure(figsize=(10, 5))
plt.plot(time, stationary_series, label='Serie de Tiempo Estacionaria')
plt.title('Ejemplo de Serie de Tiempo Estacionaria')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()
plt.grid()
plt.show()
```

Una serie de tiempo que tiene un fuerte comportamiento estacional puede ser considerada no estacionaria, ya que la media y la varianza pueden cambiar a lo largo del tiempo debido a los patrones estacionales. Sin embargo, es posible transformar una serie de tiempo estacional en una serie estacionaria mediante técnicas como la diferenciación o la eliminación de la tendencia estacional. Osea una serie que tiene estacionariedad estacional, no es estacionaria, pero se puede transformar en una serie estacionaria.



### Detectar series de tiempo no estacionarias

Para detectar si una serie de tiempo es no estacionaria, se pueden utilizar varias técnicas y pruebas estadísticas. Algunas de las más comunes incluyen:
1. **Análisis visual**: Graficar la serie de tiempo puede ayudar a identificar tendencias, patrones estacionales o cambios en la varianza que sugieran no estacionariedad
2. **Prueba de Dickey-Fuller aumentada (ADF)**: Esta prueba estadística evalúa la presencia de una raíz unitaria en la serie de tiempo, lo que indicaría no estacionariedad.
3. **Prueba de Kwiatkowski-Phillips-Schmidt-Shin (KPSS)**: Esta prueba evalúa la estacionariedad de una serie de tiempo al analizar la varianza de la serie. Si la prueba indica que la varianza no es constante, entonces la serie es no estacionaria.
4. **Prueba de Phillips-Perron (PP)**: Similar a la prueba ADF, esta prueba también evalúa la presencia de una raíz unitaria en la serie de tiempo, pero utiliza un enfoque diferente para corregir la autocorrelación y la heterocedasticidad en los errores.

### Estacionariedad Débil 
Una serie de tiempo se considera débilmente estacionaria si cumple con las siguientes condiciones:
1. La media de la serie es constante a lo largo del tiempo.
2. La varianza de la serie es constante a lo largo del tiempo.
3. La covarianza entre dos periodos de tiempo depende solo de la distancia entre esos periodos y no del tiempo específico en el que se midan.

Esto significa que la serie no muestra tendencias o patrones estacionales, y que las fluctuaciones alrededor de la media son consistentes a lo largo del tiempo. La estacionariedad débil es una condición importante para muchos modelos de series de tiempo, ya que permite realizar inferencias estadísticas y predicciones basadas en los datos históricos.



### Auto-correlación de la serie de tiempo
La auto-correlación de una serie de tiempo se refiere a la correlación entre los valores de la serie en diferentes momentos en el tiempo. En una serie de tiempo estacionaria, la auto-correlación tiende a disminuir a medida que aumenta la distancia entre los periodos de tiempo. Esto se debe a que en una serie estacionaria, los valores futuros no dependen de los valores pasados, y por lo tanto, la correlación entre los valores en diferentes momentos en el tiempo disminuye a medida que aumenta la distancia entre esos momentos. En cambio, en una serie de tiempo no estacionaria, la auto-correlación puede ser alta incluso a grandes distancias entre los periodos de tiempo, lo que indica que los valores futuros dependen de los valores pasados y que la serie no es estacionaria.
